```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_csv('flipkart.csv')
df.head()
```

```python
mode_value=df['brand'].mode()[0]
df['brand'].fillna(mode_value,inplace=True)
df.info()
```

```python
mean_value=df['retail_price'].mean()
df['retail_price'].fillna(mean_value,inplace=True)
df.info()
```

```python
mean_value=df['discounted_price'].mean()
df['discounted_price'].fillna(mean_value,inplace=True)
df.info()
```

```python
#used to delete column with null values
df.dropna(axis=1,inplace=True)
#to get information of the updated dataset
df.info()
```

```python
#converting the object into datatime datatype
df['crawl_timestamp'] = pd.to_datetime(df['crawl_timestamp'])
df['crawl_timestamp']
```

```python
df['year']=df['crawl_timestamp'].dt.year
df['year']
```

```python
df['month']=df['crawl_timestamp'].dt.month
df['month']
```

```python
sns.scatterplot(x='product_rating', y='discounted_price', data=df)
plt.title('Product Rating vs Discounted Price')
plt.show()
```

```python
# Price Analysis: Average retail and discounted prices across categories
df['main_category'] = df['product_category_tree'].str.split('>>').str[0].str.strip()
df['discount_percentage'] = ((df['retail_price'] - df['discounted_price']) / df['retail_price']) * 100

price_analysis = df.groupby('main_category').agg(
    avg_retail_price=('retail_price', 'mean'),
    avg_discounted_price=('discounted_price', 'mean'),
    avg_discount_percentage=('discount_percentage', 'mean')
).reset_index()

print(price_analysis)
top_categories = price_analysis.nlargest(10, 'avg_discounted_price')

# Set the figure size
plt.figure(figsize=(10, 6))

# Create a scatter plot
sns.scatterplot(data=top_categories, x='avg_discounted_price', y='avg_discount_percentage', hue='main_category', s=100)

# Add titles and labels
plt.title('Top 10 Categories: Avg Discounted Price vs. Avg Discount Percentage')
plt.xlabel('Average Discounted Price (â‚¹)')
plt.ylabel('Average Discount Percentage (%)')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#Market trends
# Step 1: Extract year and month for grouping
df['year_month'] = df['crawl_timestamp'].dt.to_period('M')

# Step 2: Group by year_month and calculate average metrics
trend_data = df.groupby('year_month').agg({
    'discounted_price': 'mean',
    'product_rating': 'mean'
}).reset_index()

# Step 3: Plotting the trends
plt.figure(figsize=(14, 6))

# Plot average discounted price
plt.subplot(1, 2, 1)
plt.plot(trend_data['year_month'].astype(str), trend_data['discounted_price'], marker='o', color='blue')
plt.title('Average Discounted Price Over Time')
plt.xlabel('Date')
plt.ylabel('Average Discounted Price')
plt.xticks(rotation=45)
plt.grid()

# Plot average product rating
plt.subplot(1, 2, 2)
plt.plot(trend_data['year_month'].astype(str), trend_data['product_rating'], marker='o', color='orange')
plt.title('Average Product Rating Over Time')
plt.xlabel('Date')
plt.ylabel('Average Product Rating')
plt.xticks(rotation=45)
plt.grid()

plt.tight_layout()
plt.show()
```

```python


# Step 1: Group by category and calculate average metrics
category_performance = df.groupby('main_category').agg(
    avg_discounted_price=('discounted_price', 'mean'),
    avg_product_rating=('product_rating', 'mean')
).reset_index()

# Step 2: Filter for the top 10 categories by average discounted price
top_categories = category_performance.nlargest(10, 'avg_discounted_price')

# Step 3: Create the dual-axis bar plot
plt.figure(figsize=(12, 6))  # Adjust the figure size for better spacing

# Bar plot for average discounted price
ax1 = sns.barplot(x='main_category', y='avg_discounted_price', data=top_categories, color='lightblue')
ax1.set_ylabel('Average Discounted Price (â‚¹)')
ax1.set_xlabel('Product Category')
ax1.set_title('Category Performance: Sales vs Ratings')

# Rotate x-tick labels
plt.xticks(rotation=45, ha='right')

# Create a second y-axis for product ratings
ax2 = ax1.twinx()
sns.lineplot(x='main_category', y='avg_product_rating', data=top_categories, color='red', marker='o', ax=ax2)
ax2.set_ylabel('Average Product Rating')

plt.tight_layout()
plt.show()
```

```python
# Select only the numerical columns for correlation
numerical_data = df[['retail_price', 'discounted_price', 'product_rating', 'year', 'month']]

# Compute the correlation matrix
correlation_matrix = numerical_data.corr()

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=True)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
